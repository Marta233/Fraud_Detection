## Data Preprocessing

### Class Imbalance
Initially, the dataset exhibited significant class imbalance, with the distribution of classes as follows:

| Class       | Count    |
|-------------|----------|
| Fraud       | 12268      |
| Not Fraud   | 116878  |

This imbalance posed challenges for model training, as machine learning algorithms may tend to favor the majority class.

### Applying SMOTE
To address this issue, we applied SMOTE for only training datasets, which generates synthetic examples of the minority class (fraud) to create a more balanced dataset. After applying SMOTE, the new class distribution was as follows:

| Class       | Count    |
|-------------|----------|
| Fraud       | 187004  |
| Not Fraud   | 187004  |

With the application of SMOTE, the dataset now contains an equal number of samples for both classes, allowing for a more effective model training process.

### Comparison Analysis

1. **Accuracy**:
   - The **Random Forest** model has the highest accuracy at **95.58%**, indicating that it correctly classifies a majority of instances.
   - The **Decision Tree** model performed the lowest with an accuracy of **89.19%**.

2. **Precision**:
   - The **Random Forest** model also leads in precision (**95.69%**), suggesting it has a lower rate of false positives compared to the other models.
   - The **Decision Tree** model again shows the least precision at **90.61%**.

3. **Recall**:
   - **Random Forest** and **MLP** both achieve high recall values (**95.58%** and **93.95%**, respectively), indicating they successfully identify a high proportion of true positives.
   - The **Decision Tree** model has a recall of **89.19%**, which is lower compared to the other models.

4. **F1 Score**:
   - The **Random Forest** model has the highest F1 score of **94.99%**, which balances precision and recall effectively.
   - The **Decision Tree** again has the lowest F1 score at **89.80%**.

5. **AUC**:
   - The **Random Forest** model also shows the highest AUC at **0.768564**, indicating better performance in distinguishing between classes compared to others.
   - The **Decision Tree** has the lowest AUC at **0.747311**.

### Conclusion

Based on the evaluation metrics, the **Random Forest** model outperforms the other models across all metrics, making it the best choice for this classification task. 

While the **MLP** and **LSTM** models also demonstrate strong performance, they do not surpass the Random Forest model in any of the key metrics. The **Decision Tree** model, while interpretable, shows significantly poorer performance in comparison to the ensemble methods.

In summary, for achieving the best balance between accuracy, precision, recall, and F1 score, the **Random Forest** model is recommended for deployment in practical applications.
