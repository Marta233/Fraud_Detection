## Data Preprocessing

### Class Imbalance
Initially, the dataset exhibited significant class imbalance, with the distribution of classes as follows:

| Class       | Count    |
|-------------|----------|
| Fraud       | 12268      |
| Not Fraud   | 116878  |

This imbalance posed challenges for model training, as machine learning algorithms may tend to favor the majority class.

### Applying SMOTE
To address this issue, we applied SMOTE for only training datasets, which generates synthetic examples of the minority class (fraud) to create a more balanced dataset. After applying SMOTE, the new class distribution was as follows:

| Class       | Count    |
|-------------|----------|
| Fraud       | 187004  |
| Not Fraud   | 187004  |

With the application of SMOTE, the dataset now contains an equal number of samples for both classes, allowing for a more effective model training process.

### Performance Summary:

1. **Gradient Boosting** achieved the highest accuracy (95.42%) and precision (94.22%), indicating that it correctly identified a high proportion of positive cases, though it had a lower recall (55.18%), suggesting it missed some positive cases.

2. **Logistic Regression** had the lowest precision (31.92%), indicating many of its positive predictions were false positives, but a moderate recall (61.82%).

3. **LSTM** and **MLP** models performed well with good accuracy (94.44% and 92.84%, respectively) and decent F1 scores, indicating a balanced performance between precision and recall.

4. **Decision Tree** provided a good balance with an accuracy of 89.83%, but both precision and recall were lower compared to other models, suggesting room for improvement in correctly identifying true positives.

5. **RNN** had an overall solid performance but did not stand out in any metric compared to the others.

Overall, **Gradient Boosting** showed the best performance in accuracy and precision, while the **LSTM** model had a strong overall balance in various metrics.
