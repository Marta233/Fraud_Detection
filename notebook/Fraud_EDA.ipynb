## Encoding Techniques

### 1. One-Hot Encoding
One-hot encoding was applied to the categorical features `source`, `browser`, and `sex`. This technique creates binary columns for each unique category within a feature, effectively allowing the model to interpret each category as a separate entity.
### Frequency Encoding
Frequency encoding was used for the categorical features **country** and **device_id**. This method replaces each category with its count in the dataset, which is particularly beneficial for high-cardinality features. By converting the categorical variables into their respective frequency counts, we can effectively reduce dimensionality while preserving information that may be relevant for the model.
